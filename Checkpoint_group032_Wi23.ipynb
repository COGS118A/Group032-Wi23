# Abstract 


  Machine learning has recently become an important tool in improving care for patients, by assisting in diagnosis, treatment planning, patient monitoring, and predictive analysis. Our goal is to use this tool to build a model that predicts the discharge type of patients diagnosed with diabetes. Hospital discharge type is information about why or how a patient left the hospital; could be anything from discharge to home to transfer to a new hospital to expiration. Using classification methods we will use our dataset to predict discharge type outcomes for the patients, this could potentially help hospitals to reallocate resources, reduce readmissions and improve the patient experience.

  Our dataset includes data from 130 hospitals in the United States, that were collected over 10 years. It contains over 50 features including; admission type, time in the hospital, and insulin levels, in addition, it includes hospital outcomes for each patient. The dataset was collected by extracting information from databases that satisfied specific criteria such as having diabetes and minimum length of stay being 24 hours.  We will measure the performance of our model by using various error metrics like F1 and balanced error rate. 




# Background

Diabetes is a prevalent chronic disease that poses a significant threat to human health. It is characterized by high levels of blood glucose resulting from either defective insulin secretion or impaired biological effects of insulin. The number of diabetic patients worldwide has been increasing over time, and data indicate that one in ten adults may suffer from diabetes in the future. This alarming figure highlights the urgent need for attention to combat this disease.
After being discharged from the hospital, patients may face some uncertainty about where to go for their continued recovery. If they do not receive the correct recommendations from the hospital or choose a suboptimal location, they could experience negative health consequences. This issue highlights the need for accurate prediction models that can help healthcare providers make better recommendations and ensure patients receive the best possible care.
 In recent years, machine learning has emerged as a powerful tool in the field of medical health, offering new opportunities for improving disease diagnosis and treatment. There has been significant research utilizing machine learning on the topic of diabetes, for example our dataset sourced from the Strack paper <a name="strack"></a>[<sup>[1]</sup>](#stracknote) as well as  <a name="singla"></a>[<sup>[2]</sup>](#singlanote)  and <a name="gulati"></a>[<sup>[3]</sup>](#gulatinote) Previous studies have used various machine learning algorithms such as logistic regression (LR), support vector machine (SVM), decision tree (DT CART), random forest (RF), and many more.

Charles F. Mickle conducted a 2022 study<a name="mickle"></a>[<sup>[4]</sup>](#micklenote) on predicting the type of patient discharge from multiple hospitals, in which he compared the performance of different machine learning algorithms. Specifically, he used LR, SVM, KNN, XG-Boost, and RF and evaluated their f1 scores. The study's results provide valuable insights into the performance of various machine learning algorithms for hospital discharge prediction.

One study titled "Diabetes Prediction Using Machine Learning" aims to predict whether a patient has diabetes or not based on the features provided to the machine learning model. The study uses the Pima Indians Diabetes Database, a famous dataset in the medical domain. The authors build a machine learning model, specifically a random forest model, to accurately predict the presence of diabetes in patients. The authors were inspired by prior works that used similar machine learning techniques such as decision trees and random forests. They conducted their project by analyzing and visualizing the dataset to draw insights and build a machine learning model to predict diabetes. The authors' motivation for the difference in their project was to accurately predict the presence of diabetes and draw insights from the dataset via data analysis and visualization.

Another article titled "A Machine Learning Approach to Predicting Diabetes Complications" explores the use of various machine learning algorithms to predict different complications related to diabetes. These include logistic regression (LR), support vector machine (SVM), decision tree (DT CART), random forest (RF), and k-nearest neighbors (KNN). The authors used a dataset containing various clinical features of diabetic patients, such as age, sex, BMI, and blood pressure, to predict complications such as retinopathy, neuropathy, and nephropathy. The authors found that the random forest model had the highest accuracy in predicting retinopathy, while the SVM model had the highest accuracy in predicting neuropathy and nephropathy. This study highlights the potential of machine learning in predicting and preventing complications related to diabetes, which could lead to better patient outcomes and quality of life.

In our study, we aim to compare discharge performances from various hospitals by utilizing a diverse set of machine learning models. The ultimate goal is to develop a highly accurate and efficient diabetes prediction model.


# Problem Statement

We intend to build a machine learning model to predict the discharge type of patients from a dataset of hospital admissions of patients with diabetes from the years 1999 to 2008.  There are multiple different class labels, such as sent home, transferred to another hospital, hospice, or expired, which lends itself to a multiclass prediction model.  Numerous features are associated with each patient, such as their demographics, type of admission, top 3 diagnoses, lab results, and other medications provided.  We hope to be able to predict which patients are most at risk of negative outcomes depending on the associated data, so as to better allocate hospital resources and reduce mortality for future patients with diabetes.  

# Data

[Diabetes 130-US hospitals for years 1999-2008 Data Set:](https://archive.ics.uci.edu/ml/datasets/Diabetes+130-US+hospitals+for+years+1999-2008)

 
**Dataset Info:** 100,000 observations with 50 columns

**Observation Info:**
Each observation contains patient attributes and their hospital outcomes from data collected from 1999 to 2008 at 130 US hospitals and integrated delivery networks. Relevant attributes for our model include race, gender, age, admission type, time in hospital, medical speciality of admitting physician, number of lab tests performed, HbA1c test result, diagnosis, number of medication, diabetic medications, number of outpatient, inpatient, and emergency visits in the year before hospitalization, etc.

**Critical Variables:**
Important and Relevant variables include race, age, diabetic medications, admission type, and diagnosis. Race and Admission Type are represented categorically, age and weight are represented by 10 and 25 year bins respectively, and diabetic medications are represented by a boolean. The diagnosis is represented by the first 3 digits of their ICD9 code (International Classification of Diseases, Ninth Revision, Clinical Modification).

**Data Cleaning and Transformation:**
Categorical data will need to be one-hot encoded. The Age bins will be represented by the average of the bin and then normalized to range from [0-1]. The weights column will be largely useless as most of the observations lack weight data. The diabetes medications will need to be converted from Yes/No to 0/1. Diagnoses will need to be converted from ICD9 codes to Categorical information for us to be able to understand and then transformed into a class or one-hot encoded depending on our classifier.

# Proposed Solution

Our proposed solution is to first conduct data cleaning, exploration, and preparation of the diabetes dataset.  In order to test the classification performance of how hospital patients will be discharged we will use a diverse range of algorithms such as Logistic Regression, Random Forest classifier, K-Neighbors classifier, and Multiclass Linear SVC.

**Data Cleaning and Transformation**

We chose the discharge disposition as our target column since it indicates how the patient was discharged from the hospital.  There were 29 different possible values, so to improve performance we binned the column into 4 categories, ranging from Discharged to home, Transferred to another hospital, Expired / hospice, or Unknown / null.  

Some columns had a high number of missing values, such as the weight of the patient which are null for 97% of entries, so we chose to drop this column from the training data.  

There are 9 columns which contained cardinal integer values such as age, time in hospital, number of medications, number of procedures.  We treated these as numerical columns since larger values should correlate with a larger effect. Missing values were imputed with the median for their respective columns, then normalized using StandardScaler which converts each value into a Z-Score as to not overweight one column over another.  

The remaining columns were treated as categorical since there were a relatively small number of discrete values for each category, and no obvious ordinality.  The admission source ID column has 26 possible values, ranging from Emergency Room at 57% of the data, Physican Referral at 29%, Transfer from another hospital at 3%, and smaller percentages for other types.  The race column has 6 possible values, with Caucasian composing 76% of the data, African American at 19%, with smaller percentages for other races.  There were 3 diagnosis columns consisting of a primary diagnosis, secondary, and optionally a third diagnosis.  The values were in the form of ICD9 codes which we then binned into 10 different types such as Circulatory, Respiratory, Diabetes.  The last 30 columns largely consisted of medication and lab results.  Such columns had a limited number of discrete values such as steady / up / down, which made it well suited for one hot encoding.  

After fully processing our data the feature vector has 254 dimensions, primarily because of one hot encoding.   


**Algorithms**

Our benchmark model will be logistic regression as it is one of the simplest classical machine learning models.  The 2022 paper "Early prediction of patient discharge disposition in acute neurological care using machine learning"<a name="mickle"></a>[<sup>[4]</sup>](#micklenote) established logistic regression as their benchmark which achieved an F1-score of 0.71 when classifying whether patients were discharged to home vs non-home.  

We will first try improving  binary prediction between whether the patient will be sent home or not.  The target column will be binarized based on whether the value corresponds to the discharged to home group.  We will also test multiclass performance using One-vs-Rest and a multinomial cross entropy loss function.  Multiple hyperparameter values will be explored using GridSearchCV such as the regularization value of C, different penalties such as L1 / L2 / Elastic net, and different solvers such as liblinear, saga, and newton-cg when combined with compatible regularization penalties.  For elastic net a range of penalty ratios will be explored as well, so as to balance the L1 vs L2 penalty terms. 

A decision tree classifier which will make multiclass predictions based on the binary values in one hot encodings as well as the continuous values in various features.  The random forest classifier will be another useful algorithm which forms an ensemble of decision trees on random sub-samples of our dataset. The random forest classifier can also give us feature importances, which we can then use to remove columns of low importances and reduce the dimensionality of our data.  

A K- Nearest Neighbors classifier may work well for multiclass prediction, K-Nearest Neighbors works by finding K closest neighbors and assigning class labels based on voting methods, however it is computationally expensive to perform on high dimensional data due to the “curse of dimensionality”. The curse of dimensionality implies that as the dimensionality increases it becomes harder to determine proximity due to the data points becoming more sparse. This problem can be solved by dimensionality reduction through PCA / tSNE / UMAP.     

A multiclass Linear SVC which uses one-vs-rest comparison should scale well for a relatively large dataset of approximately 100,000 data points and 100+ features after one hot encoding.  Cross validation and hyperparameter grid search will also be used to find the most optimal model for each algorithm. Alternatively, a binary SVC one-vs-one approach can also be used for multiclass classification. This approach involves training (n_classes * (n_classes - 1)) / 2 binary classifiers, where each classifier distinguishes between two classes at a time. Although this approach requires training more classifiers than the one-vs-rest approach, it can be more accurate in some cases.

# Evaluation Metrics

The dataset has multiple classes which are unbalanced, so we should use a metric which incorporates both the number of false positives and false negatives. To start out, our positive class will be a person that should be  discharged to home while our negative class will be everything else. We will weigh a false positive more heavily as that would indicate that a person is sent home when they really shouldn't have. It would be more costly to send an unwell person home (false negative) than to keep a well person in the hospital a little longer (false positive). The F1 score should be well suited, as well as the Balanced Error Rate. Moreover, we will also use Precision and Recall as a part of our metrics. Since we have multiple, unbalanced classes, this will help in interpreting our F1 score. For example, although an F1 score of 0.5 might indicate poor performance, if our positive class occurs rarely (i.e. 1% of the time), that would be considered quite good compared to guessing all-postive which yields an f1 score of 0.02 (precision = 0.01, recall= 1). We will generate confusion matrices to show the performance of the models and if there are systematic errors in the bias of the predictions.
