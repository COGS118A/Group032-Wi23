**Data Cleaning and Transformation**

We chose the discharge disposition ID as our target column since it indicates how the patient was discharged from the hospital.  There were 29 different possible values, so to improve performance we binned the column into 4 categories, ranging from discharged to Home, Transferred to another hospital, Expired / hospice, or Unknown / null.  As seen in our binning function below there were many highly specific IDs for transfers to various medical facilities, such as a specialized nursing facility, short and long term care hospital, rehab facility, ICF, among others.


````python
def discharge_disposition_id_grouping(row):
  id = row['discharge_disposition_id']
  # Home
  home = [1,7]
  if id in home:
    return 0

  # Hospice / Expired
  dead = [11, 13, 14, 19, 20, 21]
  if id in dead:
    return 1

  # Transfer to hospital
  transfer = [2,3,4,5,6,8,9,10,12,15,16,17,22,23,24,27,28,29]
  if id in transfer:
    return 2

  # Unknown 
  other = [18,25,26]
  if id in other:
    return 3

df['discharge_disposition_id_group'] = df.apply (lambda row: discharge_disposition_id_grouping(row), axis=1)
df['discharge_disposition_id_group'] 
````

Diagnosis codes were binned into major categories to improve the performance of our models.  The most common primary diagnosis category was related to circulatory issues.



````python
import math

diag_code_to_category_map = {
    0: "Circulatory",
    1: "Respiratory",
    2: "Digestive",
    3: "Diabetes",
    4: "Injury",
    5: "Musculoskeletal",
    6: "Genitourinary",
    7: "Neoplasms",
    8: "Other",
    9: "No Diagnosis"
}

def diag_to_diag_type(row, column):
  id = row[column]

  #no Diagnosis
  if id == "?":
    return 9

  try:
    id = float(id)
  except ValueError:
    return 8
 

  # Circulatory
  if (id >= 390 and id <= 459) or id == 785:
    return 0

  # Respiratory
  if (id >= 460 and id <= 519) or id == 786:
    return 1

  # Digestive
  if (id >= 520 and id <= 579) or id == 787:
    return 2

  # Diabetes 
  if math.floor(id) == 250.0 :
    return 3
  
  #Injury
  if (id >= 800 and id <=999 ):
    return 4

  # Musculoskeletal
  if (id >= 710 and id <= 739 ):
    return 5
  
  # Genitourinary
  if (id >= 580 and id <= 629 ) or id == 788:
    return 6
  
  # Neoplasms

  if (id >= 140 and id <= 239):
    return 7
  #other
  return 8

df['diag_1_type'] = df.apply (lambda row: diag_to_diag_type(row, 'diag_1_code3'), axis=1)
df['diag_1_category'] = df['diag_1_type'].map(diag_code_to_category_map)
df['diag_1_category'].value_counts()
````

# Evaluation Metrics

The dataset has multiple classes which are unbalanced, so we should use a metric which incorporates both the number of false positives and false negatives. To start out, our positive class will be a person that should be  discharged to home while our negative class will be everything else. We will weigh a false positive more heavily as that would indicate that a person is sent home when they really shouldn't have. It would be more costly to send an unwell person home (false negative) than to keep a well person in the hospital a little longer (false positive). The F1 score should be well suited, as well as the Balanced Error Rate. Moreover, we will also use Precision and Recall as a part of our metrics. Since we have multiple, unbalanced classes, this will help in interpreting our F1 score. ### Model 1: Logistic Regression

Logistic Regression is a linear model for classification known for using the sigmoid or logistic function to form binary predictions.  Multi-class predictions can be achieved through One-vs-Rest classification which compares one class against all the others.  Multinomial logistic regression can also be performed for multi-class prediction by using the softmax function to generate class probabilities for each data point.  The primary hyperparameters are the regularization constant C, the penalty terms L1 / L2 / Elastic-net, the L1 ratio in the case of Elastic-net, and the solver algorithm.  

Preliminary model selection was performed on a binary prediction task, where the only classes were whether the patient was discharged to home or not.  Grid search was run across a range of C values from 0.01 to 10, penalties L1, L2 and Elasticnet, as well as different solvers which were compatible for the given penalties.  It was found that Elasticnet was most effective at an F1 score of 0.788, but only marginally compared to L2 and L1 at 0.787 with C values of 0.1.  We concluded that the penalties and solvers had negligible effect on the final performance of the model.  

To better approach the multi-dimensional nature of the dataset we chose to focus on the multi-class prediction task.  Multi-class logistic regression models were explored using GridSearchCV with C values ranging from 0.01 to 10.  We chose to only use the Elasticnet penalty with saga solver, and L1 ratios 0.25, 0.5 and 0.75.  The best performing model used a L1 ratio of 0.25 and C of 0.01, resulting in an F1 score of 0.702.  The corresponding Grid Search code can be seen below:

Regularization is used to prevent overfitting by adding a penalty term to the SVM objective function, which encourages the model to have smaller weights and simpler decision boundaries.
We tested four different values of C: 0.01, 0.1, 1.0, and 10.0. Lower values of C produce a smoother decision boundary and reduce overfitting, while higher values of C increase the model's sensitivity to the training data and may lead to overfitting.

We used 5-fold cross-validation to assess the performance of the model with each value of C. For each fold, we trained the model on 80% of the training data and validated it on the remaining 20%. We computed several evaluation metrics to compare the performance of the models, including accuracy, F1 score, and balanced accuracy.

The results of the hyperparameter tuning are summarized in the table below:

c.JPG


From the table, we can see that the best performance was achieved with C=1.0, which produced an F1 score of 0.7125, an accuracy of 0.7125, and a balanced accuracy of 0.3788. The model with C=1.0 had the highest balanced accuracy among all the tested values of C, indicating that it achieved a good balance between sensitivity and specificity.

We then fit the SVM model with C=1.0 using the full training set, and evaluated its performance on the test set. The model took approximately 89 minutes to fit and predict the data. The evaluation metrics on the test set were as follows:

F1 score: 0.7125

Accuracy: 0.7125

Balanced accuracy: 0.3788

Precision: 0.7125

Recall: 0.7125

Overall, our results suggest that moderate regularization with C=1.0 is preferable to either weak or strong regularization for this dataset and problem. This finding is consistent with the intuition that overly complex models may overfit to the training data, while overly simple models may underfit and fail to capture the underlying patterns in the data.






````python
from sklearn.svm import SVC

svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', verbose=1)

svm_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                               ('model', svm_model)])

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.3, random_state=42)

scores = cross_val_score(svm_pipeline, X_train, y_train, scoring='f1_micro', verbose=1, n_jobs=-1)

# Preprocessing of training data, fit model 
svm_pipeline.fit(X_train, y_train)

# Preprocessing of validation data, get predictions
preds = svm_pipeline.predict(X_test)

# Evaluate the model
f1 = f1_score(y_test, preds, average='micro')
print('F1:', f1)
acc = accuracy_score(y_test, preds)
print('accuracy_score', acc)
bacc = balanced_accuracy_score(y_test, preds) 
print('balanced accuracy', bacc)
prec = precision_score(y_test, preds, average='micro') 
print('precision', prec)
rec = recall_score(y_test, preds, average='micro')
print('recall', rec)
````

#### Bootstrap 



In an effort to improve the training speed of the model, a bootstrap method was applied, which resampled the training data with replacement and trained the SVM model on each resampled dataset. The number of iterations was set to 100, and the mean and standard deviation of the F1-score, accuracy score, and balanced accuracy score were calculated. However, the performance of the bootstrap method was not better than the initial SVM model, achieving an F1-score mean of 0.677 with a standard deviation of 0.004. Further experimentation may be necessary 

Therefore, it was concluded that the SVM model was not the best fit for the dataset, and alternative machine learning algorithms should be explored.

````python
from sklearn.utils import resample

# define the number of iterations
n_iterations = 100

# define lists to store the f1 and accuracy scores for each iteration
f1_scores = []
accuracy_scores = []
bacc_scores = []

# perform bootstrap
for i in range(n_iterations):
    # bootstrap sample
    X_train_resampled, y_train_resampled = resample(X_train, y_train, random_state=i)
    
    # fit the model on the resampled data
    svm_pipeline.fit(X_train_resampled, y_train_resampled)
    
    # predict on the test data
    preds = svm_pipeline.predict(X_test)
    
    # calculate evaluation metrics
    f1 = f1_score(y_test, preds, average='micro')
    f1_scores.append(f1)
    
    acc = accuracy_score(y_test, preds)
    accuracy_scores.append(acc)
    
    bacc = balanced_accuracy_score(y_test, preds)
    bacc_scores.append(bacc)

# calculate the mean and standard deviation of the scores
f1_mean = np.mean(f1_scores)
f1_std = np.std(f1_scores)

acc_mean = np.mean(accuracy_scores)
acc_std = np.std(accuracy_scores)

bacc_mean = np.mean(bacc_scores)
bacc_std = np.std(bacc_scores)

# print the results
print('F1 Score - Mean:', f1_mean, 'Std Dev:', f1_std)
print('Accuracy Score - Mean:', acc_mean, 'Std Dev:', acc_std)
print('Balanced Accuracy Score - Mean:', bacc_mean, 'Std Dev:', bacc_std)
````

In this paper, we used scikit learns implementation of the random forest classifier and used a grid search, with 5-fold cross-validation, to search for the optimal hyperparameters. ### Model 4: Gradient Boosting

Gradient Boosting is an ensemble learning method that combines numerous weak learners and trains them sequentially while iteratively correcting errors made by previous learners.  The final prediction of the model is an aggregated prediction of all learners. 
In this paper, we have implemented the sci-kit learn version of Gradient Learning and performed a grid search to determine the best hyperparameters. We used the prepocessing as explained above and created a pipeline for the gradient boosting. Then grid search was done across the parameters; (100,200,300) for number of estimators, (0.1,0.05,0.01) for learning rate and (3,4,5) for maximum depth. Grid search concluded that the best parameters for the dataset were; a learning rate of 0.05, a maximum depth of 5, and a number of estimators of 300. After these findings we fit the model with the best parameters. The mean cross-validated score of the best estimator was 72.22%.The F1 score of the model was 0.72 consistent with the other models.  




````python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.3, random_state=42)


# Preprocessing 
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler()),
    ])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

# Pipieline
gb_pipe = Pipeline(steps=[('preprocessor', preprocessor), ('gb', gb)])

# Parameter grid 
param_grid = {
    'gb__n_estimators': [100, 200,300],
    'gb__learning_rate': [0.1, 0.05,0.01],
    'gb__max_depth': [3, 4,5],
}


grid_search = GridSearchCV(gb_pipe, param_grid=param_grid, cv=3,  n_jobs= -1)

# Fitting the model
grid_search.fit(X_train, y_train)

#best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

#best estimator 
best_gb_pipeline = grid_search.best_estimator_

#best estimator 
best_gb_pipeline.fit(X_train, y_train)

# Prediction
y_pred = best_gb_pipeline.predict(X_test)

#F1 score
f1 = f1_score(y_test, y_pred, average='micro')

print('F1:', f1)
print('Best parameters:', best_params)
print("Best score: {:.2f}%".format(best_score * 100))
````

### Model 5: Multi-Layer Perceptron

Our last model explored various structures for the Multi-Layer Perceptron (MLP), a simple neural network which can be computed on a CPU.  It is a supervised learning algorithm which takes a set of features X and an output target y, with fully connected hidden layers in between which contain a number of neurons with weights between the input and output layers.  Gradient descent is performed through backpropagation, where the loss gradients are propagated backwards to the neurons' weights according to their respective partial derivatives.  MLPs can approximate an arbitrary function due to the non-linearity introduced by their activation function, which is typically a Rectified Linear Unit (ReLU).  

We set up a MLPClassifier using Adam as the solver, an alpha L2 regularization term set to 1e-5, and learning rate set to adaptive.  Grid search was conducted over networks with hidden layers 5 neurons wide and 2 layers deep to 200 neurons wide and 3 layers deep.  The best MLPClassifier model had hidden layer sizes of 100 neurons by 2 layers, with an F1 score of 0.710.

The code to generate this model was similar as before, using the same pipeline and preprocessing steps, only with the model MLPClassifier from SKLearn, and a param grid of different hidden layer sizes.


````python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(solver='adam', alpha=1e-5, learning_rate='adaptive', 
                    hidden_layer_sizes=(5, 2), random_state=42, verbose=0, max_iter=10000)

mlp_pipe = Pipeline(steps=[('preprocessor', preprocessor),
                            ('mlp', mlp)
                             ])

param_grid = {
     'mlp__hidden_layer_sizes' : [(5, 2), (5, 3), (10, 2), (10,3), (20,2), (20,3), (100,2), (100,3), (200,2), (200,3)],
     'mlp__early_stopping': [True],
    }

gscv_mlp = GridSearchCV(mlp_pipe, param_grid, scoring='f1_micro', cv=5, n_jobs=-1, verbose=4)

gscv_mlp.fit(X_train, y_train)

clf = gscv_mlp.best_estimator_

preds = clf.predict(X_test)

mlp_pipe.fit(X_train, y_train)

preds = mlp_pipe.predict(X_test)

f1 = f1_score(y_test, preds, average='micro')
print('F1:', f1)
````

Our major findings were about the models, and the feature importance, they are equally important for predicting the discharge type of patients:

The most optimal model in the multi-class prediction task was the Gradient Boosting Classifier (GBC) which achieved an F1 score of 0.722 with a learning rate of 0.05, a maximum depth of 5, and number of estimators of 300.  This corroborates our expection given the tabular nature of our dataset and relatively small scale compared to the vast training sets for large deep neural networks.  Boosting models are known to perform well on tabular data as shown by XGBoost winning many Kaggle competitions.    

We also expected the GBC to outperform the Random Forest (RF) model's F1 score of 0.708 since Boosting is a more advanced form of ensemble learning.  However we were surprised that dimensionality reduction did not significantly improve the model much after selecting the top 100 columns based on feature importance. The Multi-Layer Perceptron performed moderately well with an F1 score of 0.710. However it is a very simple neural network with 2 hidden layers of 100 neurons wide.
